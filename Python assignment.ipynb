```python
input2.rename(columns = {'uid' : 'User ID','name':'Name'},inplace =True)
```

```python
newdf = pd.merge(input1,input2)
```

```python
# newdf['Total_Score'] = newdf['total_statements']+ newdf['total_reasons']
```

```python
newdf['Team Name'].unique()
```

```python
def teamnamecorrection(word):
    word = str(word)
    return word.title()
```

```python
newdf['temp_Team_Name'] = newdf['Team Name'].apply(teamnamecorrection)
```

```python
newdf
```

```python
temp = newdf.groupby('temp_Team_Name').agg('mean')
```

```python
# temp = temp.drop(columns = ['User ID','S No'])
```

```python
temp
```

```python
temp['TotalAvg'] = temp['total_statements'] + temp['total_reasons'] 
```

```python
temp
```

```python
firstoutput = temp.sort_values('TotalAvg',ascending = False)
```

```python
firstoutput.insert(0,'Rank',range(1,1+firstoutput.shape[0]))
```

```python
firstoutput.drop('TotalAvg',axis = 1)
```

```python
newdf2 = newdf[['Name','User ID','total_statements','total_reasons']]
```

```python
newdf2['totalScore'] = newdf2['total_statements'] + newdf2['total_reasons']
```

```python
newdf2
```

```python
def small(word):
    word = str(word)
    res = ''
    for i in word:
        if i.isalpha():
            res += i.lower()
    return res
```

```python
newdf2['smallname'] = newdf2['Name'].apply(small)
```

```python
newdf2['TotalScore'] = newdf2['total_statements'] + newdf2['total_reasons']
```

```python
newdf2
```

```python
newdf2 = newdf2.sort_values(by=['TotalScore','smallname'],ascending = (False,True))
```

```python
newdf2
```

```python
output2 = newdf2[['Name','User ID','total_statements','total_reasons']]
```
